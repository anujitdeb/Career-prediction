```python
#This part is for Train and Test

from sklearn.model_selection import train_test_split

x = df.iloc[:,df.columns != 'Output']
y = df.iloc[:,df.columns == 'Output']

print("Here is the value of x ------------>>>>>>>>>\n")
print(x)
print("\n\n")
print("Here is the value of y ------------>>>>>>>>>\n")
print(y)
print("\n\n")
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.15)

#Here I select 15% of my dataset for testing purpose and remaining 85% of my dataset for traning purpose
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
# Here the implementation of Algorithm and Model building

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

# Here I train the model

model.fit(xtrain, ytrain.values.ravel()) # Here I use Ravel function for avoiding DataConversionWarning
```

```python
# Here I predict the output

prediction = model.predict(xtest)

print(prediction)
```

```python
from sklearn.metrics import  accuracy_score

accuracy = accuracy_score(prediction, ytest)

print("Therefore the Accuracy is : ", accuracy)
```
